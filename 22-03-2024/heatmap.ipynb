```python
df = pd.read_csv("./ponto de partida/clicks.csv")
df.sample(3)
```

```python
sns.scatterplot(
    df,
    x="X Position",
    y="Y Position",
    c="k"
)
```

```python
image = cv2.imread('./fundo.jpeg')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)

df_escalado = df.copy()
df_escalado["X Position"] = df_escalado["X Position"]*0.85
df_escalado["Y Position"] = df_escalado["Y Position"]*0.9

sns.scatterplot(
    df_escalado,
    x="X Position",
    y="Y Position",
    c="k", s=1
)

sns.kdeplot(
    df_escalado,
    x="X Position",
    y="Y Position",
    lw=0.01, fill=True, alpha=.5, cmap="rocket"
)

plt.xlim (0, image.shape[1])
plt.ylim (image.shape[0], 0)

plt.axis('off')
plt.tight_layout()
plt.savefig("heatmap.png", dpi=1200)
plt.show()

```
